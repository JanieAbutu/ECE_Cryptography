```python
# ElGamal Example

# Public domain parameters
p = 23        # prime modulus
g = 7         # generator

# Alexandre's keys
dA = 10                         # private key
eA = pow(g, dA, p)              # public key

# Bernard's keys
eB = 15                         # given public key

print("1️## Verify Alexandre’s data:")
print("eA =", eA)               # should be 13

# Encryption: Alexandre sends x=12 to Bernard, k=7
x = 12
k = 7
r = pow(g, k, p)
y = (x * pow(eB, k, p)) % p

print("\n2️## Ciphertext from Alexandre to Bernard:")
print("(r, y) =", (r, y))

# Decryption: Alexandre receives (r, y) = (20, 11)
r_recv, y_recv = 20, 11
rdA = pow(r_recv, dA, p)
inv_rdA = pow(rdA, -1, p)       # modular inverse
x_plain = (y_recv * inv_rdA) % p

print("\n3️## Decrypt Bernard’s message:")
print("Plaintext =", x_plain)

# Find Bernard’s private key dB such that 7^dB = 15 (mod 23)
dB = None
for i in range(1, p):
    if pow(g, i, p) == eB:
        dB = i
        break

print("\n4️# Bernard’s private key:")
print("dB =", dB)
```
