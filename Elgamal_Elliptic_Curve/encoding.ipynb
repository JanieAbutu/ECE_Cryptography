```python
# ======================================= 
# ECC ElGamal Letter Encoding with Offset (Exact Output)
# =======================================

p = 89
a = 22
b = 28

# Modular square root
def sqrt_mod(a, p):
    for x in range(p):
        if (x*x) % p == a:
            return x
    return None

# Encode letter with offsets (A=0, B=1,...)
def encode_letter_exact(ch):
    m = ord(ch.upper()) - 65   # A=0, B=1, ..., Z=25
    offset = 0
    first_attempt_y = None
    while offset < 200:
        x = (m + offset) % p
        rhs = (x*x*x + a*x + b) % p
        y = sqrt_mod(rhs, p)
        if y is not None:
            if offset == 0:
                # Original number corresponds to a valid point
                print(f"The letter {ch} corresponds to the number {m}. "
                      f"Now, the point ({x}, {y}) belongs to E. Thus, {ch} ←→ (({x}, {y}), {offset}).")
            else:
                # Original number did NOT belong; needed offset
                print(f"The letter {ch} corresponds to the number {m}. "
                      f"There are no points in E with x = {m}, {m+1}, ..., {m+offset-1}. "
                      f"However, ({x}, {y}) ∈ E. Thus, {ch} ←→ (({x}, {y}), {offset}).")
            return (x, y), offset
        offset += 1
    raise ValueError("Encoding failed.")

# Encode the message
message = "PARIS"
encoded = []
for ch in message:
    Pm, off = encode_letter_exact(ch)
    encoded.append((Pm, off))
```
